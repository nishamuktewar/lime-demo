```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import savefig
from sklearn.ensemble import IsolationForest
```

```python
# Generate some data
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=400, centers=4,
                       cluster_std=0.60, random_state=0)
```

```python
# Creating pandas dataframe from numpy array
dataset = pd.DataFrame({'A':X[:,0],'B':X[:,1]})
dataset[:5]
```

```python
class_names=['group 1', 'group 2', 'group 3', 'group 4']
```

```python
plt.scatter(dataset['A'], dataset['B'], s=40, cmap='viridis');
```

```python
rng = np.random.RandomState(123)
# training the model
clf = IsolationForest(max_samples=100, contamination =0.1, random_state=rng, behaviour='new')
clf.fit(dataset)

# predictions
pred = clf.predict(dataset)
```

```python
plt.scatter(dataset['A'], dataset['B'], c=pred, s=40, cmap='viridis');
```

The predicted values are -1 (outlier) and 1 (inlier)

```python
pred
```

And at the max you have scores, higher the score the more of an inlier it is

```python
pred_scores = clf.decision_function(dataset)
```
